Nonogramy

W niniejszym projekcie zajęłam się problemem rozwiązywania nonogramów przy pomocy algorytmu genetycznego, a także przy użyciu PSO (roju). 
Rozwiązanie było testowane najpierw na małych nonogramach 5x5, a później na większych.

```python
import pygad
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
pyramid_reg_poz = [[1], [2], [3], [4], [5]]
pyramid_reg_pion = [[5], [4], [3], [2], [1]]

curr_reg = [pyramid_reg_poz, pyramid_reg_pion]
```

Przetestujemy rozwiązanie najpierw za pomocą algorytmu genetycznego, w tym celu importujemy potrzebne biblioteki i definiujemy reguły pionowe i poziome nonogramu, który będziemy rozwiązywać.

```python
gene_space = [0,1]
sol_per_pop = 1000
num_parents_mating = 500
num_generations = 50
keep_parents = 20

parent_selection_type = "sss"
crossover_type = "single_point"
mutation_type = "random"
mutation_percent_genes = 4
```

Następnie stawiamy parametry dla alogrytmu. Naszym rozwiązaniem jest tablica zer i 
jedynek spełniających zdefiniowane wcześniej reguły. 0 oznacza kratkę niezamalowaną, 1 - zamalowaną.
Wybrano dużą populację (1000 osobników) aby wynik uzyskać po minięciu jak najmniejszej liczby generacji. 

```python
def create_chunks(list_name, n):
    for i in range(0, len(list_name), n):
        yield list_name[i:i + n]


def get_columns(chunks):
    columns = []
    for i in range(len(chunks[0])):
        col = []
        for j in range(len(chunks)):
            col.append(chunks[j][i])
        columns.append(col)
    
    return columns
```

Definiujemy funkcje pomocnicze, rozdzielające rozwiązanie na wiersze i kolumny

```python
def check_row(row):
    length = 0
    rul = []
    for gene in row:
        if gene == 1:
            length += 1
        elif length != 0 and gene == 0:
            rul.append(length)
            length = 0

    if length != 0:
        rul.append(length)

    return rul

def check_column(column):
    length = 0
    rul = []
    for gene in column:
        if gene == 1:
            length += 1
        elif length != 0 and gene == 0:
            rul.append(length)
            length = 0

    if length != 0:
        rul.append(length)

    return rul
```

A także wydzielające reguły z wierszy i kolumn chromosomu. Następnie, w funcji fitness, są one porównywane z regułami w nonogramie.

```python
def fit(genes, solution_idx):
    reg_poz, reg_pion = curr_reg
    dl = len(reg_poz)
    szer = len(reg_pion)
    chunks = list(create_chunks(genes, szer))
    columns = get_columns(chunks)
    
    points = 0
    for i in range(len(chunks)):
        row = chunks[i]
        rul_row = check_row(row)
        reg_i_row = reg_poz[i]
        if rul_row == reg_i_row:
            points += 1
        elif sum(rul_row) != sum(reg_i_row):
            points -= 0.5

    for j in range(len(columns)):
        column = columns[j]
        rul_col = check_column(column)
        reg_j_col = reg_pion[j]
        if rul_col == reg_j_col:
            points += 1
        elif sum(rul_col) != sum(reg_j_col):
            points -= 0.5

    return points
```

Za każdą spełnioną regułę (rul_col == reg_j_col) jest przydzielany 1 punkt. Maksymalna liczba punktów w naszym nonogramie to 10.

```python
def plot(array):
    img = np.array(array, dtype=np.float64)
    img = np.stack([img, img, img], axis=2)
    img = np.abs(img - 1)
    plt.imshow(img)
    plt.show()
```

Definiujemy funkcję, która wyświetli nasze rozwiązanie (listę zer i jedynek) w postaci obrazu.

```python
def run(reg):
    reg_poz, reg_pion = reg
    dl = len(reg_pion)
    szer = len(reg_poz)
    num_genes = dl * szer
    fitness_function = fit
    max_points = dl + szer
    criteria = "reach_" + str(max_points)

    # inicjacja algorytmu z powyzszymi parametrami wpisanymi w atrybuty
    ga_instance = pygad.GA(gene_space=gene_space,
                           num_generations=num_generations,
                           num_parents_mating=num_parents_mating,
                           fitness_func=fitness_function,
                           sol_per_pop=sol_per_pop,
                           num_genes=num_genes,
                           parent_selection_type=parent_selection_type,
                           keep_parents=keep_parents,
                           crossover_type=crossover_type,
                           mutation_type=mutation_type,
                           mutation_percent_genes=mutation_percent_genes,
                           stop_criteria=[criteria])

    # uruchomienie algorytmu
    ga_instance.run()

    # podsumowanie: najlepsze znalezione rozwiazanie (chromosom+ocena)
    solution, solution_fitness, solution_idx = ga_instance.best_solution()
    chunks = list(create_chunks(solution, szer))
    sum_chunks = list([sum(chunk) for chunk in chunks])
    print("Parameters of the best solution : ")
    for chunk in chunks:
        print(chunk)
    print(f"Fitness value of the best solution = {solution_fitness}")

    # wyswietlenie wykresu: jak zmieniala sie ocena na przestrzeni pokolen
    ga_instance.plot_fitness()

    plot(chunks)
```

Funkcja run robi kilka rzeczy: inicjuje algorytm z podanymi wcześniej atrybutami, uruchamia go, wyświetla najlepsze znalezione rozwiązanie razem z wykresem oceny na przestrzeni pokoleń.
Wyświetla również nasze rozwiązanie w postaci obrazu (wcześniej zdefiniowana funkcja plot).

```python
start = time.time()
run(curr_reg)

end = time.time()
seconds = (end - start)
print(seconds)
```

Wypróbujmy teraz rozwiązywanie nonogramów za pomocą PSO (roju).

```python
import pyswarms as ps
from pyswarms.utils.plotters import plot_cost_history
import matplotlib.pyplot as plt
```

```python
def fit_swarm(genes):
    reg_poz, reg_pion = curr_reg
    dl = len(reg_poz)
    szer = len(reg_pion)
    points = 0.0

    chunks = list(create_chunks(genes, szer))
    columns = get_columns(chunks)

    for i in range(len(chunks)):
        row = chunks[i]
        rul_row = check_row(row)
        reg_i_row = reg_poz[i]
        if rul_row == reg_i_row:
            points += 1.0

    for j in range(len(columns)):
        column = columns[j]
        rul_col = check_column(column)
        reg_j_col = reg_pion[j]
        if rul_col == reg_j_col:
            points += 1.0

    return points


def f(swarm):
    return [(-1) * fit_swarm(particle)
            for particle in swarm]
```

```python
def print_sol(sol):
    chunks = create_chunks(sol, len(curr_reg[0]))

    lst = []
    for row in chunks:
        print(row)
        lst.append(row)

#     print(lst)
    plot(lst)
```

```python
curr_reg = [pyramid_reg_poz, pyramid_reg_pion] # schodki1 5x5

options = {'c1': 0.5, 'c2': 0.3, 'w':0.9, 'k':2, 'p':1}

n_genes = len(curr_reg[0]) * len(curr_reg[1])
print(n_genes)

start = time.time()
optimizer = ps.discrete.BinaryPSO(n_particles=10, dimensions=n_genes,
options=options)
best_cost, best_pos = optimizer.optimize(f, iters=5000, verbose=True)

end = time.time()
t = end - start
print(t)

cost_history = optimizer.cost_history
plot_cost_history(cost_history)
plt.show()

print_sol(best_pos)
```
